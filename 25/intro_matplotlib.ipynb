### Gráfico de Dispersão

Pode-se usar uma string de formatação para se plotar o gráfico de dispersão utilizando a matplotlib. Isso é feito na forma **[marcador][estilo_linha][cor]**. Por exemplo, pode-se criar uma linha tracejada preta com **'--k'** ou um gráfico de dispersão vermelho com **'or'**. Detalhes sobre a string de formatação podem ser vistos [aqui](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

Neste exemplo é calculada também a **correlação**, que é mostrada na figura através de um elemento texto.

```python
%matplotlib inline
n_linhas = 20
corr = fb['high'][:n_linhas].corr(fb['low'][:n_linhas])
plt.plot('high', 'low', 'or', data=fb.head(n_linhas))

x_max = plt.xlim()[1]  # valor máximo do eixo x do gráfico
y_min = plt.ylim()[0]  # valor mínimo do eixo y do gráfico

plt.text(x_max, y_min, f'Correlação = {corr:.2f}', ha='right', va='bottom')
```

### Histogramas

Histogramas são úteis para avaliar a ocorrência de determinados valores em uma distribuição.

A aparência do histograma é fortemente influenciada pelo número de bins.


```python
%matplotlib inline
terremotos = pd.read_csv(f"/content/drive/MyDrive/{caminho_dados}/earthquakes.csv")
terremotos_ml = terremotos.query('magType == "ml"').mag
fig, axes = plt.subplots(3, 1)
for ax, bins in zip(axes, [7, 15, 35]):
    ax.hist(terremotos_ml, bins=bins)
    ax.set_title(f'bins param: {bins}')
plt.subplots_adjust(hspace=0.5)
```

#### Figura com um **grid** de gráficos

```python
fig = plt.figure(figsize=(8, 8))
gs = fig.add_gridspec(3, 3)
top_left = fig.add_subplot(gs[0, 0])
mid_left = fig.add_subplot(gs[1, 0])
top_right = fig.add_subplot(gs[:2, 1:])
bottom = fig.add_subplot(gs[2,:])
```

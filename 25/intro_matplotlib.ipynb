### Gráfico de Dispersão

Pode-se usar uma string de formatação para se plotar o gráfico de dispersão utilizando a matplotlib. Isso é feito na forma **[marcador][estilo_linha][cor]**. Por exemplo, pode-se criar uma linha tracejada preta com **'--k'** ou um gráfico de dispersão vermelho com **'or'**. Detalhes sobre a string de formatação podem ser vistos [aqui](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

Neste exemplo é calculada também a **correlação**, que é mostrada na figura através de um elemento texto.

```python
%matplotlib inline
n_linhas = 20
corr = fb['high'][:n_linhas].corr(fb['low'][:n_linhas])
plt.plot('high', 'low', 'or', data=fb.head(n_linhas))

x_max = plt.xlim()[1]  # valor máximo do eixo x do gráfico
y_min = plt.ylim()[0]  # valor mínimo do eixo y do gráfico

plt.text(x_max, y_min, f'Correlação = {corr:.2f}', ha='right', va='bottom')
```

### Histogramas

Histogramas são úteis para avaliar a ocorrência de determinados valores em uma distribuição.

A aparência do histograma é fortemente influenciada pelo número de bins.


```python
%matplotlib inline
terremotos = pd.read_csv(f"/content/drive/MyDrive/{caminho_dados}/earthquakes.csv")
terremotos_ml = terremotos.query('magType == "ml"').mag
fig, axes = plt.subplots(3, 1)
for ax, bins in zip(axes, [7, 15, 35]):
    ax.hist(terremotos_ml, bins=bins)
    ax.set_title(f'bins param: {bins}')
plt.subplots_adjust(hspace=0.5)
```

## Personalização

É possível tornar os gráficos mais atraentes e informativos com recursos da matplotlib. Dentre as principais personalizações estão:

- **Título e rótulos**: adicionar títulos e rótulos aos eixos com a função `title()` e `xlabel()` e `ylabel()`.
- **Legenda**: adicionar legendas às linhas com a função `legend()` e personalizar sua localização e formato.
- **Eixos**: personalizar os limites dos eixos com `xlim()` e `ylim()`, definir os rótulos dos ticks do eixo com `xticks()` e `yticks()`, e definir a escala dos eixos com `xscale()` e `yscale()`.
- **Estilo de linhas e marcadores**: personalizar a cor, estilo e largura das linhas com `color()`, `linestyle()` e `linewidth()`, e personalizar os marcadores com `marker()` e `markersize()`.
- **Fundo**: adicionar grades aos gráficos com a função `grid()`, alterar as cores de fundo das figuras e dos eixos com `facecolor()`, e adicionar retângulos coloridos com `axvspan()` ou `axhspan()`.
- **Subplots** : criar subplots em uma figura com a função `subplot()`, personalizar o tamanho e as margens dos subplots com `subplots_adjust()`, e adicionar títulos e rótulos aos subplots individualmente.
- **Figura**: personalizar o tamanho da figura com `figure(figsize=())`, adicionar um título à figura com `suptitle()`, e salvar a figura em diferentes formatos de arquivo com a função `savefig()`.
- **Textos e anotações**: adicionar textos e anotações em pontos específicos do gráfico com `text()` e `annotate()`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados aleatórios
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Criar figura com subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Plotar dados no primeiro subplot
axs[0].plot(x, y1, color='blue', linewidth=2, linestyle='--', label='Seno')
axs[0].scatter(x, y1, marker='o', color='blue')
axs[0].set_xlabel('Eixo X', fontdict={'fontsize': 12, 'fontweight': 'bold'})
axs[0].set_ylabel('Eixo Y', fontdict={'fontsize': 12, 'fontweight': 'bold'})
axs[0].set_title('Gráfico de Seno', fontdict={'fontsize': 16, 'fontweight': 'bold'})
axs[0].set_xlim(0, 10)
axs[0].set_ylim(-1.2, 1.2)
axs[0].grid()
axs[0].legend(loc='upper right')

# Plotar dados no segundo subplot
axs[1].plot(x, y2, color='green', linewidth=2, linestyle='-.', label='Cosseno')
axs[1].scatter(x, y2, marker='o', color='green')
axs[1].set_xlabel('Eixo X', fontdict={'fontsize': 12, 'fontweight': 'bold'})
axs[1].set_ylabel('Eixo Y', fontdict={'fontsize': 12, 'fontweight': 'bold'})
axs[1].set_title('Gráfico de Cosseno', fontdict={'fontsize': 16, 'fontweight': 'bold'})
axs[1].set_xlim(0, 10)
axs[1].set_ylim(-1.2, 1.2)
axs[1].grid()
axs[1].legend(loc='upper right')

# Adicionar linhas de referência no segundo subplot
axs[1].axhline(y=0, color='red', linewidth=0.8)
axs[1].axvline(x=np.pi/2, color='red', linewidth=0.8, linestyle='--')

# Adicionar título à figura
fig.suptitle('Dois Gráficos em uma Figura', fontsize=20, fontweight='bold')

# Ajustar margens dos subplots
fig.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, hspace=0.4)

# Adicionar seta com texto no primeiro subplot
axs[0].annotate('Máximo', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 0.8),
                arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12)
plt.show()
```
